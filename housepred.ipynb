```python
def convertrange(x):
    temp=x.split('-')
    if len(temp)==2 :
        return float(temp[0])+float(temp[1])/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft']=df['total_sqft'].apply(convertrange)
```

```python
def remove_outliers(df1):
    df1_output=pd.DataFrame()
    for key ,subdf1 in df1.groupby('location'):
        m=np.mean(subdf1.price_per_sqft)
        st=np.std(subdf1.price_per_sqft)
        gen_df1=subdf1[(subdf1.price_per_sqft > (m-st)) & (subdf1.price_per_sqft <= (m+st))]
        df1_output=pd.concat([df1_output,gen_df1],ignore_index=True)
    return df1_output
df=remove_outliers(df)
df.describe()
```

```python

def bhk_outlier(df1):
    exclude_indices = []  # Use a list instead of np.array for efficiency

    for location, location_df1 in df1.groupby('location'):
        bhk_stats = {}

        for bhk, bhk_df1 in location_df1.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df1.price_per_sqft),
                'std': np.std(bhk_df1.price_per_sqft),
                'count': bhk_df1.shape[0]
            }

        for bhk, bhk_df1 in location_df1.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)  # Fix variable name

            if stats and stats['count'] > 5:  # Ensure bhk-1 exists in bhk_stats
                exclude_indices.extend(bhk_df1[bhk_df1.price_per_sqft < stats['mean']].index.values)

    return df1.drop(exclude_indices, axis='index')

# Assuming df is your DataFrame
df = bhk_outlier(df)
df
```

```python
y=df["price"]

X=df.drop(columns=["price"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),  # Encoding categorical feature
    remainder='passthrough'  # Keep the remaining numerical columns as they are
)
```

```python
df.shape
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```
